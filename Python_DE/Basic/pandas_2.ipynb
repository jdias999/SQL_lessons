```python
import pandas as pd
from datasets import load_dataset


dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'], errors='coerce') #transformar em datetime
```

```python
df['job_posted_date'] = df['job_posted_date'].dt.date

df.head(3)
```

```python
df.sort_values(by='job_posted_date', ascending=False, inplace=True)
df.head(3)
```

```python
df.dropna(subset=['salary_year_avg'], inplace=True) #para remover valores nan
df.head(3)
```

```python
df['job_title_short'].value_counts().head(3) #contar valores de cada trabalho
```

```python
#exemplo de groupby
df.groupby('job_title_short')['salary_year_avg'].median()
```

```python
df.groupby('job_title_short')['salary_year_avg'].agg(['min', 'max'])
```

```python
#agora, alguns exercícios

us_jobs = df[df['job_country'] == 'United States'] #separa apenas jobs dos eua
us_jobs.groupby('job_title_short')['salary_year_avg'].mean().sort_values(ascending=False).head(5) # faz o groupby, e coloca em ordem decrescente
```

```python
us_jobs.groupby('job_title_short').size().sort_values(ascending=False) #pega a contagem de cada trabalho
```

```python
#exemplo final

us_jobs.groupby('job_title_short')['salary_year_avg'].agg(['median', 'min', 'max', 'count']).sort_values(by='median', ascending=False)
```
