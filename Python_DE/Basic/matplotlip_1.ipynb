```python
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt  #chamando o matplotlip

dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

```python
df['job_posted_month'] = df['job_posted_date'].dt.month

monthly_counts = df['job_posted_month'].value_counts() #faz os meses virarem index

monthly_counts = monthly_counts.sort_index() #aqui, os meses ja são index
monthly_counts 
```

```python
plt.plot(monthly_counts.index, monthly_counts)
```

```python
job_counts = df['job_title_short'].value_counts().head(3)

plt.bar(job_counts.index, job_counts)
plt.show()
```

```python
job_counts = df['job_title_short'].value_counts().head(6).sort_values(ascending=True)

job_counts
```

```python
plt.barh(job_counts.index, job_counts) 
plt.show()
```

```python
job_counts = df['job_title_short'].value_counts()

plt.bar(job_counts.index, job_counts)  
plt.ylabel('Count of Postings') 
plt.title('Job Postings by Job Title')
plt.xticks(rotation=45, ha='right') #para ficar legível
plt.show()
```

```python
#agora, fazer o mesmo usando pandas

job_counts = df['job_title_short'].value_counts()

job_counts.plot(kind='bar')
plt.xlabel('')
plt.ylabel('Count of Postings')
plt.title('Job Postings by Job Title')
plt.xticks(rotation=45, ha='right') 
plt.show()
```

```python
#exemplo final

job_salary = df.groupby('job_title_short')['salary_year_avg'].median().sort_values()

job_salary.plot(kind='barh')
plt.xlabel('Salary ($USD)')
plt.ylabel('')
plt.title('Median Salary by Job Title') 
plt.show()
```
